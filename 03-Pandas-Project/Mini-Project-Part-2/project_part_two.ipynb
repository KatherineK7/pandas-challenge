```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csvpath = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
bootcamp = pd.read_csv(csvpath)
```

```python
# Take only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
bootcamp_extract = bootcamp.iloc[:,[0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
bootcamp_extract.head()
```

```python
# Change reading 0 and 1 to No and Yes, respectively
bootcamp_extract = bootcamp_extract.replace({0.0: 'No',1.0: 'Yes'})
bootcamp_extract.head()
```

```python
# Extract rows for only those who attended a bootcamp
bootcamp_attendees = bootcamp_extract.loc[bootcamp_extract['AttendedBootcamp'] == 'Yes']
bootcamp_attendees.head()
```

```python
# Create DataFrame of the different boot camps that had a significant number of attendees
well_attended = pd.DataFrame(bootcamp_attendees['BootcampName'].value_counts())
well_attended.head()

well_attended.reset_index(inplace=True)
well_attended.columns = ['BootcampName','Attendees']

well_attended = well_attended.loc[well_attended['Attendees'] >= 20]
well_attended
```

```python
# Count how many attendees of each bootcamp would recommend the bootcamp
would_recommend = pd.DataFrame(bootcamp_attendees.loc[bootcamp_attendees['BootcampRecommend'] == 'Yes', 'BootcampName'].value_counts())

would_recommend.reset_index(inplace=True)

would_recommend.columns = ['BootcampName','Recommendations']
would_recommend
```

```python
# Merge the two created data frames on the name of tbe bootcamp
merged = pd.merge(well_attended, would_recommend, on='BootcampName')
merged
```

```python
# Calculate percentage of eac bootcamp's students who are recommenders
#(recommendations/attendees*100) for x in list

percentage = round((merged['Recommendations']/merged['Attendees'])*100,2)
percentage

merged['Percentage Recommended'] = percentage
merged.head()

merged = merged.sort_values(by=['Percentage Recommended'], ascending=False)
```

```python
# Export to excel and remove index

merged.to_excel("output/BootcampOutputPart2.xlsx", index=False)
```
