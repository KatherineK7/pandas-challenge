```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csvfile = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
bootcamp = pd.read_csv(csvfile)
```

```python
# Inspect all columns
bootcamp.head()
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
bootcamp_extract = bootcamp.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
bootcamp_extract.head()
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
bootcamp_extract = bootcamp_extract.replace({0.0:"No", 1.0:"Yes"})
bootcamp_extract.head()
```

```python
# Calculate total number of respondents in survey
total_respondents = len(bootcamp_extract)
total_respondents
```

```python
# Extract rows corresponding only to people who attended a bootcamp
bootcamp_attendees = bootcamp_extract.loc[bootcamp_extract['AttendedBootcamp']== 'Yes',:]
bootcamp_attendees.head()
```

```python
# Calculate average age of attendees
average_age = bootcamp_attendees['Age'].mean()
average_age
# Calculate how many people attended a bootcamp
total_attendees = len(bootcamp_attendees)
total_attendees
# Calculate how many attendees hold degrees
degree_holders = len(bootcamp_attendees.loc[(bootcamp_attendees["SchoolDegree"]== "bachelor's degree") \
                                        | (bootcamp_attendees["SchoolDegree"]== "master's degree (non-professional)") \
                                           | (bootcamp_attendees["SchoolDegree"]== "professional degree (MBA, MD, JD, etc.)") \
                                            | (bootcamp_attendees["SchoolDegree"]== "associate's degree") \
                                                | (bootcamp_attendees["SchoolDegree"]== "high school diploma or equivalent (GED)")\
                                                   | (bootcamp_attendees["SchoolDegree"]== "Ph.D."), :])
                                           

# Count number of attendees who self-identify as male; female; or are of non-binary gender identification

male_attendees = len(bootcamp_attendees.loc[bootcamp_attendees['Gender'] == 'male', :])


female_attendees = len(bootcamp_attendees.loc[bootcamp_attendees['Gender'] == 'female', :])


nonbinary_attendees = total_attendees - male_attendees - female_attendees
nonbinary_attendees

# Calculate percentage of respondents who attended a bootcamp

prct_attended = round(total_attendees / total_respondents * 100, 2)
prct_attended

# Calculate percentage of respondents belonging to each gender
male_respondents = len(bootcamp_extract.loc[bootcamp_extract['Gender'] == 'male',:])

prct_male_respondents = round(male_respondents / total_respondents * 100, 2)
prct_male_respondents

female_respondents = len(bootcamp_extract.loc[bootcamp_extract['Gender'] == 'female',:])
female_respondents

prct_female_respondents = round(female_respondents / total_respondents * 100, 2)
prct_female_respondents

nonbinary_respondents = total_respondents - male_respondents - female_respondents
prct_nonbinary_respondents = round(nonbinary_respondents / total_respondents * 100, 2)
prct_nonbinary_respondents

# Calculate percentage of attendees with a school degree

prct_attendees_degrees = round(degree_holders / total_attendees * 100, 2)
prct_attendees_degrees


# Calculate average post-bootcamp salary
average_post_bootcamp_salary = round(bootcamp_attendees['BootcampPostSalary'].mean(),2)
average_post_bootcamp_salary
```

```python
# Create a new table consolodating above calculations


output_dict = {"Average Age of Attendees": [average_age], "Total Attendees": [total_attendees],"Male Attendees": [male_attendees],\
               "Female Attendees":[female_attendees], "Nonbinary Attendees": [nonbinary_attendees],\
               "Percent Respondents Attended": [prct_attended], "Percent Male Respondents": [prct_male_respondents], \
              "Percent Female Respondents": [prct_female_respondents], "Percent Nonbinary Respondents": [prct_nonbinary_respondents],\
               "Pecent Attendees with Degrees": [prct_attendees_degrees], "Average Post-Bootcamp Salary": [average_post_bootcamp_salary]}

output_df = pd.DataFrame(output_dict)
output_df

```
